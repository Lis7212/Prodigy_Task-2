```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
titanic_df = pd.read_csv("Titanic.csv")

# Display the first few rows of the dataset
titanic_df.head()
```

```python
# Check for missing values
print(titanic_df.isnull().sum())

# Visualize missing values
sns.heatmap(titanic_df.isnull(), cmap='viridis', yticklabels=False, cbar=False)
plt.show()
```

```python
sns.countplot(x='Survived', data=titanic_df, palette='coolwarm')
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic_df, palette='coolwarm')
plt.title('Survival by Gender')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic_df, palette='coolwarm')
plt.title('Survival by Class')
plt.show()
```

```python
sns.histplot(data=titanic_df, x='Age', bins=30, kde=True, hue='Survived', palette='coolwarm')
plt.title('Age Distribution by Survival')
plt.show()
```

```python
sns.countplot(x='SibSp', data=titanic_df, palette='coolwarm')
plt.title('Siblings/Spouses Count')
plt.show()
```

```python
sns.countplot(x='Parch', data=titanic_df, palette='coolwarm')
plt.title('Parents/Children Count')
plt.show()
```
